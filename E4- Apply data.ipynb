#Create a lambda function that will capitalize strings. and Capitalize both Mjob and Fjob

```
cap = lambda x:x.capitalize()
df['Mjob'] = df['Mjob'].apply(cap)
df['Fjob'] = df['Fjob'].apply(cap)
df
```

```

# capi = lambda x: x.str.capitalize()

# sdf.select(capi('Mjob')).show()
```

```
 
@pandas_udf('string') 
def capitalize(s1): 
  # Here we are using s1.'str'.capitalize() as  
  # s1 is a pandas Series object and it  
  # doesn't contain capitalize() method.  
  # It is a string method, that's why we have written 
  # s1.str.capitalize() 
  return (s1.str.capitalize()) 
  
sdf.select(capitalize('Mjob')).show()
sdf.select(capitalize('Fjob')).show()
sdf.withColumn('Mjob', upper(sdf['Mjob'])).show()
```

#Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)


```
# def majority (x):
#   return True if x>17 else False

majority = lambda x: x > 17
df['legal_drinker'] = df['age'].apply(majority)
df
```

```
sdf.withColumn('legal_drinker', expr('age')>17).show()
```

#Multiply every number of the dataset by 10.

```
times10 = lambda x: x*10 if type(x) is int else x
df.applymap(times10)
```
